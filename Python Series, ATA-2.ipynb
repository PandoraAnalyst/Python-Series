## 3. Comparison, Logical, Membership Operators

### -- COMPARISON OPERATORS
These are used to determine the equality of two or more variables. It is used for both string and number/integer data types. 
They are used for comparisons and to solve logical problems. 

1. #### The Equal to (==) & Not Equal to (!=) Operators.

```python
### Examples of (==)
20 == 30
```

```python
### Examples of (!=)
50 != 30
```

```python
### Examples of both
x = 20
y = 15
z = 20

print(x != y)
print(x == z)
```

2. #### The Less Than (<), Greater Than (>), Less Than or Equal To (<=), and Greater Than or Equal To (>=) Operators.

```python
### Examples of (<)
20 < 30
```

```python
### Examples of (>)
30 > 20
```

```python
### Examples of (<=)
20 <= 40
```

```python
### Examples of (>=)
50 >= 16
```

### -- LOGICAL OPERATORS
These operators are often combined with Comparison operators. There are 3 of them: and, or, and not.

#### 1. The 'and' Operator

```python
## Examples of 'and' operator.
(20 > 50) and (30 > 10)
```

```python
## Examples of 'and' operator.
(60 > 50) and (30 > 10)

## NOTE: Both conditions are true here, so it returns TRUE
```

```python
## Examples of 'and' operator.
('Vanilla' > 'Chocolate') and (80 > 60)

## This returns TRUE because both conditions are true. Vanilla is greater than Chocolate because 
## 'V' is no 22 in the alphabets while 'C' is no 3, and 80 is greater than 60.
```

#### 2. The 'or' Operator

```python
## Examples of 'OR' operator.
(20 > 50) or (30 > 10)
```

```python
## Examples of 'OR' operator.
('Vanilla' > 'Chocolate') or (10 > 60)

## This returns TRUE because one condition is true. Vanilla is greater than Chocolate because 
## 'V' is no 22 in the alphabets while 'C' is no 3.
```

#### 3. The 'not' Operator

```python
## Examples of 'not' operator
not(50 > 10)
```

```python
## Examples of 'not' operator
not(10 < 0)
```

### -- MEMBERSHIP OPERATORS
These are used to check if something, whether it's a value or a string is within another value or string or sequence.

```python
## Examples
ice_cream = 'I love chocolate ice creams'

'love' in ice_cream
```

```python
scoops = [1,2,3,4,5]

2 in scoops
```

```python
scoops = [1,2,3,4,5]

6 in scoops
```

```python
scoops = [1,2,3,4,5]

6 not in scoops
```

## 4. If, Elif, and Else Statements
These statements are conditional statements that are used to run a body of code depending on the result of a comparison, logical or membership operations. If something is true, the code stops there, otherwise, it'll run the Elif body. If Elif result is also false, then it runs the body of the Else condition.

```python
## Example of 'if' statement

if 25 > 10:
    print('Worked!')
```

```python
## Example of 'if else' statement
if 25 < 10:
    print('Worked!')
else:
    print('it does not work')
```

```python
## Example of 'if elif else' statement
if 25 < 10:
    print('Worked!')
elif 25 < 30:
    print('elif Worked!')
else:
    print('it does not work')

## NOTE: there can be many Elif statements in a check
```

```python
## Example of many 'elif' statement in a conditional check
if 25 < 10:
    print('Worked!')
elif 25 < 20:
    print('elif Worked!')
elif 25 < 21:
    print('elif 2 Worked!')
elif 25 < 41:
    print('elif 3 Worked!')
else:
    print('it does not work')
```

```python
## Example using logical operators
if (25 < 10) or (1 < 3):
    print('Worked!')
elif 25 < 20:
    print('elif Worked!')
```

```python
## Exmaples of writing if else statement in one line

print('it worked') if 10>30 else print('it did not work')
```

```python
## Nested 'if' statements

if (25 < 10) or (1 < 3):
    print('Worked!')
    if 10 > 5: 
        print('Nested statement worked')
elif 25 < 20:
    print('elif Worked!')
```
