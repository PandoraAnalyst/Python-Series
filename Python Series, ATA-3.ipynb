## 5. For Loops
A 'For' Loop is used to iterate over a sequence which could be a list, tuple, an array, string, or even a dictionary.



```python
## Examples

integers = [1,2,3,4,5,6,7]

for number in integers:
    print(number)
```

```python
for number in integers:
    print(number + number)
```

```python
## 'For' Loops handles a dictionary a little bit differently

ice_creams_dict = {'Name': 'Pandora', 'weekly intake': 5, 'favorite ice creams': ['MCC','Chocolate']}
```

```python
for cream in ice_creams_dict.values():
    print(cream)
```

```python
## We can also pull two at once from a dict item using a for loop

for key, value in ice_creams_dict.items():
    print(key, '->', value)
```

#### Nested For Loops
This is simply a For loop within another For Loop.

```python
## Example for nested for loops

flavors = ['Vanilla', 'Chocolate', 'Cookie Dough']
toppings = ['Oreos', 'Marshmallows', 'Hot Fudge']
```

```python
### Looping through the list above using nested for loops

for one in flavors:
    for two in toppings:
        print(one, 'topped with', two)
```

## 6. While Loops
The 'While' loop is used to iterate over a block of code as long as the test condition is true. The diff between a 'For' loop and a 'While' loop is a FOR loop is going to iterate over the entire sequence regardless of a condition, but the WHILE loop is only going to iterate over that sequence as long as a specific condition is met. Once that condition is not met, the code is gonna stop and it's not going to iterate over the rest of the sequence. 
The 'while' loop is sometimes used in conjuction with a 'break' statement. This statement stops the loop even if the while condition is true. The 'Continue' statement is also used with While loops; when this statement is triggered, it basically rejects all remaining statements in the current iteration of the loop and then go to the next iteration.



```python
## Example for 'while' loop

number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
## Example for 'while' loop using 'Break' statement

number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
## Example for 'while' loop using 'Else' statement

number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
## Example for 'while' loop using 'Continue' statement

number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
else:
    print('No longer < 5')

## When the number is equal to 3, because of continue, it skips 3 and goes back to the top and continues the loop
```
